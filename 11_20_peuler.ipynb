Largest Product in a Grid / Problem 11

In the 20 x 20 grid below, four numbers along a diagonal line have been marked in red.



The product of these numbers is 26 x 63 x 78 x 14 = 1788696.

What is the greatest product of four adjacent numbers in the same direction (up, down, left, right, or diagonally) in the 20 x 20 grid?

```python
# Matrix

matrix = [
    [8, 2, 22, 97, 38, 15, 0, 40, 0, 75, 4, 5, 7, 78, 52, 12, 50, 77, 91, 8],
    [49, 49, 99, 40, 17, 81, 18, 57, 60, 87, 17, 40, 98, 43, 69, 48, 4, 56, 62, 0],
    [81, 49, 31, 73, 55, 79, 14, 29, 93, 71, 40, 67, 53, 88, 30, 3, 49, 13, 36, 65],
    [52, 70, 95, 23, 4, 60, 11, 42, 69, 24, 68, 56, 1, 32, 56, 71, 37, 2, 36, 91],
    [22, 31, 16, 71, 51, 67, 63, 89, 41, 92, 36, 54, 22, 40, 40, 28, 66, 33, 13, 80],
    [24, 47, 32, 60, 99, 3, 45, 2, 44, 75, 33, 53, 78, 36, 84, 20, 35, 17, 12, 50],
    [32, 98, 81, 28, 64, 23, 67, 10, 26, 38, 40, 67, 59, 54, 70, 66, 18, 38, 64, 70],
    [67, 26, 20, 68, 2, 62, 12, 20, 95, 63, 94, 39, 63, 8, 40, 91, 66, 49, 94, 21],
    [24, 55, 58, 5, 66, 73, 99, 26, 97, 17, 78, 78, 96, 83, 14, 88, 34, 89, 63, 72],
    [21, 36, 23, 9, 75, 0, 76, 44, 20, 45, 35, 14, 0, 61, 33, 97, 34, 31, 33, 95],
    [78, 17, 53, 28, 22, 75, 31, 67, 15, 94, 3, 80, 4, 62, 16, 14, 9, 53, 56, 92],
    [16, 39, 5, 42, 96, 35, 31, 47, 55, 58, 88, 24, 0, 17, 54, 24, 36, 29, 85, 57],
    [86, 56, 0, 48, 35, 71, 89, 7, 5, 44, 44, 37, 44, 60, 21, 58, 51, 54, 17, 58],
    [19, 80, 81, 68, 5, 94, 47, 69, 28, 73, 92, 13, 86, 52, 17, 77, 4, 89, 55, 40],
    [4, 52, 8, 83, 97, 35, 99, 16, 7, 97, 57, 32, 16, 26, 26, 79, 33, 27, 98, 66],
    [88, 36, 68, 87, 57, 62, 20, 72, 3, 46, 33, 67, 46, 55, 12, 32, 63, 93, 53, 69],
    [4, 42, 16, 73, 38, 25, 39, 11, 24, 94, 72, 18, 8, 46, 29, 32, 40, 62, 76, 36],
    [20, 69, 36, 41, 72, 30, 23, 88, 34, 62, 99, 69, 82, 67, 59, 85, 74, 4, 36, 16],
    [20, 73, 35, 29, 78, 31, 90, 1, 74, 31, 49, 71, 48, 86, 81, 16, 23, 57, 5, 54],
    [1, 70, 54, 71, 83, 51, 54, 69, 16, 92, 33, 48, 61, 43, 52, 1, 89, 19, 67, 48]
]


def up_product(matrix, row, column):
    if row > 16:
        return 0
    else:
        return matrix[row][column] * matrix[row+1][column] * matrix[row+2][column] * matrix[row+3][column]        

def down_product(matrix, row, column):
    if row < 2:
        return 0
    else:
        return matrix[row][column] * matrix[row-1][column] * matrix[row-2][column] * matrix[row-3][column]

def right_product(matrix, row, column):
    if column > 16:
        return 0
    else: 
        return matrix[row][column] * matrix[row][column+1] * matrix[row][column+2] * matrix[row][column+3]
    
def left_product(matrix, row, column):
    if column < 2:
        return 0
    else:
        return matrix[row][column] * matrix[row][column-1] * matrix[row][column-2] * matrix[row][column-3]

def diagonally_down_right(matrix, row, column):
    if (row > 16) or (column > 16):
        return 0
    else:
        return matrix[row][column] * matrix[row+1][column+1] * matrix[row+2][column+2] * matrix[row+3][column+3]

def diagonally_down_left(matrix, row, column):
    if (row > 16) or (column < 2):
        return 0
    else:
        return matrix[row][column] * matrix[row+1][column-1] * matrix[row+2][column-2] * matrix[row+3][column-3]
    
def largest_product_in_a_grid():

    greatest = 0
    maybe = 0

    for row in range(20):
        for column in range(20):
            maybe = max(
                up_product(matrix, row, column),
                down_product(matrix, row, column),
                right_product(matrix, row, column),
                left_product(matrix, row, column),
                diagonally_down_right(matrix, row, column),
                diagonally_down_left(matrix, row, column)
            )

            if maybe > greatest:
                greatest = maybe
    
    return greatest
```

Highly Divisible Triangular Number / Problem 12

The sequence of triangle numbers is generated by adding the natural numbers. So the 7th triangle number would be 1 + 2 + 3 + 4 + 5 + 6 + 7 = 28. 

The first ten terms would be: 1, 3, 6, 10, 15, 21, 28, 36, 45, 55,...

Let us list the factors of the first seven triangle numbers:



We can see that 28 is the first triangle number to have over five divisors.

What is the value of the first triangle number to have over five hundred divisors?

```python
def triangle_number(n:int):

    return n * (n + 1) // 2

def find_divisor_number(n):

    if n <= 1:
        return 1
     
    remaining = n
    divisor_count = 1
    p = 2

    while p * p <= remaining:
        if remaining % p == 0:
            exponent = 0
            while remaining % p == 0:
                remaining //= p
                exponent += 1
            divisor_count *= (exponent + 1)
        p += 1 if p == 2 else 2  
    
    if remaining > 1:
        divisor_count *= 2
    return divisor_count


def highly_divisible_triang_num(threshold: int = 500):

    n = 1

    while 1 == 1:
        
        if n % 2 == 0:
            divisor_total = find_divisor_number(n // 2) * find_divisor_number(n + 1)
        else:
            divisor_total = find_divisor_number(n) * find_divisor_number((n + 1) // 2)

        if divisor_total > threshold:
            return triangle_number(n)
        n += 1
```

```python
if __name__ == "__main__":
    print(highly_divisible_triang_num(500))
```
